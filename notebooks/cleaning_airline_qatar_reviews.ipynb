```python
aircraft_pattern = patterns = [
    r'Boeing \d{3}(-\d+)?',  # Matches 'Boeing 787', 'Boeing 787-777', etc.
    r'A3\d{2} (neo)?',         # Matches 'A319', 'A321neo', etc.
]
route_pattern = r'([A-Za-z]+(?:-[A-Za-z]+)*\s*(?:to|–|-)\s*[A-Za-z]+(?:-[A-Za-z]+)*(?:\s+via\s+[A-Za-z]+(?:-[A-Za-z]+)*)*)'
month_year_pattern = r'([A-Za-z]+ \d{4})'
```

```python
def check_row_for_valid_type(row, valid_types):
    for val in row:
        if val in valid_types:
            return val  # Return the first match found
    return None  # If no match is found, return None
```

```python
def check_for_aircraft(row, patterns):
    for val in row:
        if isinstance(val, str):  # Ensure the value is a string
            for pattern in patterns:
                if re.search(pattern, val):  # If pattern is found in the value
                    return re.search(pattern, val).group()  # Return the matching pattern
    return None  # Return None if no match is found
```

```python
def check_for_route(row):
    # Iterate through each value in the row
    for val in row:
        if isinstance(val, str):  # Check if the value is a string
            if ' to ' in val.lower() or ' via ' in val.lower():  # Check for 'to' or 'via'
                return val  # Return the matching value
    return None  # Return None if no match is found
```

```python
def extract_route(text):
    if text is None:  # Handle None case explicitly
        return None  # You can return an empty string or 'No route found' if you prefer
    match = re.search(route_pattern, text)  # Search for a matching route pattern
    if match:
        return match.group(0)  # Return the matched route
    return None  # If no route is found, return None
```

```python
def extract_month_year(text):
    # Check if the data is not a string (if it's NaN, a number, list, etc.)
    if not isinstance(text, str):  # If it's not a string, return None
        return None
    
    # If text is a valid string, apply the regex search
    match = re.search(month_year_pattern, text)  # Search for the pattern
    if match:
        return match.group(0)  # Return the matched "Month Year" format
    return None  # If no match, return None
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    text = re.sub(r'[^\w\s]', '', text.lower())  # Lowercase and remove punctuation
    text = ' '.join([word for word in text.split() if word not in stop_words])  # Remove stop words
    text = ' '.join([lemmatizer.lemmatize(word) for word in text.split()])  # Lemmatize
    return text
```

```python
def standardize_route(route):
    # Split the route by ' to ' and ' via '
    route = route.lower()
    parts = route.split(' to ')
    if len(parts) == 2:
        origin = parts[0].strip()
        destination_and_via = parts[1].strip().split(' via ')
        destination = destination_and_via[0].strip()
        via = destination_and_via[1].strip() if len(destination_and_via) > 1 else ''
        
        # Join the parts in the desired format
        new_route = f"{origin}-{via}-{destination}".strip('-')
        return new_route
    return route  # Return the original route if it doesn't match the expected format
    
```

```python
def split_route(route):
    if pd.isna(route):
        return pd.Series([np.nan, np.nan, np.nan])
    
    # Replace '--' and split by spaces and hyphens
    route = route.replace('--', '-').replace(' via ', ' - ').replace(' - ', '-')
    airports = route.split('-')
    
    source = airports[0].strip()
    destination = airports[-1].strip()
    via = [airport.strip() for airport in airports[1:-1]] if len(airports) > 2 else []
    
    return pd.Series([source, destination, via])
```

```python
def map_airports(row):
    source_code = airport_mapping.get(row['Source'].lower(), row['Source'])
    destination_code = airport_mapping.get(row['Destination'].lower(), row['Destination'])
    via_codes = [airport_mapping.get(airport.lower(), airport) for airport in row['Via']]
    
    return pd.Series([source_code, destination_code, via_codes])
```

```python
def split_via(via_list):
    if isinstance(via_list, list) and via_list:
        # Flatten the list and split on '//' or any other delimiter
        return [item for sublist in via_list for item in sublist.split('//')]
    return []
```

```python
def uppercase_via(via_list):
    if isinstance(via_list, list) and via_list:
        # Split on '//' and convert to uppercase
        return [item.upper() for sublist in via_list for item in sublist.split('//')]
    return []
```

```python
df_airline_reviews['Route'] = df_airline_reviews['Route'].astype(str)
df_airline_reviews['Route'] = df_airline_reviews['Route'].apply(standardize_route)
df_airline_reviews[['Source', 'Destination', 'Via']] = df_airline_reviews['Route'].apply(split_route)
df_airline_reviews[['Source', 'Destination', 'Via']] = df_airline_reviews.apply(map_airports, axis=1)
code_columns = ['Source', 'Destination']
df_airline_reviews[code_columns] = df_airline_reviews[code_columns].apply(
    lambda col: col.where(col.isna(), col.str.upper())
)
df_airline_reviews['Via'] = df_airline_reviews['Via'].apply(split_via)
df_airline_reviews['Via'] = df_airline_reviews['Via'].apply(uppercase_via)


```

```python
def clean_and_split_aircraft(aircraft_str):
    if pd.isna(aircraft_str) or aircraft_str.strip().upper() == 'UNKNOWN':
        return np.nan
    # Normalize separators and split
    cleaned_str = aircraft_str.replace('/', ',').replace(' AND ', ',').replace(' and ', ',')
    return [a.strip().upper() for a in cleaned_str.split(',')]
```
